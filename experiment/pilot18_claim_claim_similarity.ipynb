```python
# reading claims, perspectives and their paraphrases 
file = "/Users/daniel/ideaProjects/perspective/data/perspective_paraphrases/Pilot-16-perspective-paraphrase-verification_raw_all.csv"
import csv

rows = []
with open(file, mode='r') as fileReader:
    readCSV = csv.reader(fileReader, delimiter=',')
    for row in readCSV: 
        rows.append(row)
```

```python
print(len(rows))
```

```python
import json 

# convert this data into per claim: 
claim_to_stuff_map = {}
for x in rows[1:]: 
#     print(x)
    claim_id = x[0]
    claim_title = x[1]
    perspective_id = x[2]
    p_title = x[3]
    paraphrase_perspective = x[4]
    para_title = x[5]
    vote_yes = x[6]
    vote_no = x[7]
    if claim_id not in claim_to_stuff_map: 
        claim_to_stuff_map[claim_id] = {"claim_title": claim_title, "claim_id": claim_id, "perspectives": [], "concat": ""}
    claim_to_stuff_map[claim_id]["perspectives"].append({"p_id": perspective_id, "text": p_title})
    if vote_yes > vote_no: 
        claim_to_stuff_map[claim_id]["perspectives"].append({"p_id": paraphrase_perspective, "text": para_title})
print(len(claim_to_stuff_map))
for k in claim_to_stuff_map.keys(): 
    claim_to_stuff_map[k]["concat"] = " ".join([x["text"] for x in claim_to_stuff_map[k]["perspectives"]])
claim_to_stuff_strings = [json.dumps(x) for x in claim_to_stuff_map.values()]
```

```python
# create json for each row and write the output on disk: 
with open('/Users/daniel/ideaProjects/perspective/data/perspective_paraphrases/Pilot-16-perspective-paraphrase-verification_raw_all.json', "w") as file:  
    file.write("\n".join(claim_to_stuff_strings))
```

```python
# find top claim pairs 
similar_claims = []
for x in claim_to_stuff_map.keys(): 
    claim_id = claim_to_stuff_map[x]["claim_id"]
    claim_title = claim_to_stuff_map[x]["claim_title"]
#     print(claim_title)
    out = es.get_claims(claim_title)
    for y in out[0:20]: 
        (concat2, claim_title2, id2, score) = y
        similar_claims.append({"claim1_title": claim_title, 
                               "claim1_id": claim_id, 
                               "claim2_title": claim_title2, "claim2_id": id2, "score": score})
```

```python
# rank the most-similar pairs 
similar_claims = sorted(similar_claims, key=lambda x: -x["score"])
```
