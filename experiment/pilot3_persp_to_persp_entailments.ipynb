```python
import json 

id_to_perspective_map = {}
with open("/Users/daniel/ideaProjects/perspective/data/perspectives.json") as file:  
    perspectives = json.load(file)
    for p in perspectives: 
        id_to_perspective_map[p["id"]] = p["title"]
print(len(id_to_perspective_map))
```

```python
p_ids = id_to_perspective_map.keys()

id_to_id_scores = {}
for pair in perspective_pairs: 
    id1 = pair["id1"]
    id2 = pair["id2"]
    if id1 != id2: 
        if id1 in id_to_id_scores: 
            id_to_id_scores[id1].append(pair)
        else: 
            id_to_id_scores[id1] = [pair]

            
id_to_id_scores2 = {}
for pair in perspective_pairs: 
    id1 = pair["id1"]
    id2 = pair["id2"]
    if id1 != id2: 
        if (id1,id2) in id_to_id_scores2: 
            id_to_id_scores2[(id1,id2)].append(pair)
        else: 
            id_to_id_scores2[(id1,id2)] = [pair]

        if (id2,id1) in id_to_id_scores2: 
            id_to_id_scores2[(id2,id1)].append(pair)
        else: 
            id_to_id_scores2[(id2,id1)] = [pair]

            
            
```

```python
import math 

def normalize(a): 
    return str(math.floor(a * 100) / 100.0)

def printProb(p): 
    return normalize(p[0]) + ", " + normalize(p[1]) + ", " + normalize(p[0] + p[1])
```

```python
# sorted according to top entailment scores 
for id1 in list(id_to_id_scores.keys())[0:40]: 
    perspective1 = id_to_perspective_map[id1]
    pairs = id_to_id_scores[id1]
    pairs.sort(key=lambda x: x["prob"][0], reverse=True)
    pairs = pairs[0:10]
#     if len(pairs) > 5: 
    print("------")
    print(f"P ({id1}): " + perspective1)
    for x in pairs: 
        print(f"Related P ({x['id2']}): {id_to_perspective_map[x['id2']]} ({printProb(x['prob'])})")
```

```python
# sorted according to top entailment + contradictions scores 
for id1 in list(id_to_id_scores.keys())[0:40]: 
    perspective1 = id_to_perspective_map[id1]
    pairs = id_to_id_scores[id1]
    pairs.sort(key=lambda x: x["prob"][0] + x["prob"][1], reverse=True)
    pairs = pairs[0:10]
#     if len(pairs) > 5: 
    print("------")
    print("P: " + perspective1)
    for x in pairs: 
        print(f"Related P ({x['id2']}): {id_to_perspective_map[x['id2']]} ({printProb(x['prob'])})")
```

```python
# sorted according to top entailment scores, in two directions 
for id1 in list(id_to_id_scores.keys())[0:40]: 
    perspective1 = id_to_perspective_map[id1]
    print("------")
    print("P: " + perspective1)
    pairs = []
    for id2 in id_to_perspective_map.keys(): 
        if (id1, id2) in id_to_id_scores2: 
            scores = id_to_id_scores2[(id1, id2)]
            final_score = [x["prob"][0] for x in scores]
            # print(final_score)
            # print(sum(final_score) / len(final_score))
            pairs.append([id2, sum(final_score) / len(final_score)])
    pairs.sort(key=lambda x: x[1], reverse=True)
    pairs = pairs[0:10]
    for x in pairs: 
        print(f"Related P({x[0]}): {id_to_perspective_map[x[0]]} ({normalize(x[1])})")
```
